```python
import json
import csv

categoris = []
with open(".data/US_category_id.json") as file:
    us_cat = json.load(file)["items"]
    categories = [ [item["id"], item["snippet"]["title"]] for item in us_cat ]

with open("./categories.csv", "w") as output_categories:
    csv_writer = csv.writer(output_categories)
    csv_writer.writerow(["id", "category_name"])
    csv_writer.writerows(categories)
```
