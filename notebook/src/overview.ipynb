```python


tags = []
with open(".data/tags.txt") as file:
    raw_tags = file.readlines()
    for raw_tag in raw_tags:
        tags.append(raw_tag.strip().split("#")[1])

print(len(tags))

tags = set(tags)

tags = [ [tag, random.randint(1, 10000)] for tag in tags]

with open("./tags.csv", "w") as output_categories:
    csv_writer = csv.writer(output_categories)
    csv_writer.writerow(["name", "count"])
    csv_writer.writerows(tags)
```

```python


all_df = pd.DataFrame(columns=["name", "count", "country"])

def get_tags(data, country):
    
    names = []
    for raw_tag in data["tags"]:
        tag_names = raw_tag.split("|")
        tag_names = [ item.lower().strip() for item in raw_tag.split("|") if item]

        names.extend(tag_names)

    np_names = np.array(names)
    tags, counts = np.unique(names, return_counts=True)
    
    return dict(zip(tags, counts))
    

for country in ["US", "CA"]:
    raw_data = pd.read_csv(f".data/{country}_youtube_trending_data.csv")
    tags = get_tags(data=raw_data, country=country)
    
    data = { "name": list(tags.keys()), "count": list(tags.values())}

    df = pd.DataFrame(data=data)
    df["country"] = country
    
    all_df = pd.concat([all_df, df])
    
    


```

```python
all_df.to_csv("./data_tags.csv", sep=",")
```
