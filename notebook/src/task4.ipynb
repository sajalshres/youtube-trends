```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/boxplot1.csv')
data
```

```python
sns.boxplot(x=data.f) 
```

```python
sns.boxplot(x=data.Character)
```

```python
sns.distplot(x=data.Character,kde=True)
```

```python
sns.violinplot(x=data.Character,inner="box")
```

```python
characters = []
for _, row in data.iterrows():
    temp = [ row.Character for item in range(1, row.f + 1)]
    characters.extend(temp)
    
```

```python
vplot = sns.violinplot(x=characters,inner="box")
```
